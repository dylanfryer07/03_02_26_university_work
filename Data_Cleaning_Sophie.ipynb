```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/Users/sophie.taylor/Git Hub/03_02_26_university_work/Forward_Feast_Survey.csv")
```

```python
data.head()
```

```python
missing_per_column = data.isnull().sum()
#print("Missing per column:\n", missing_per_column)

total_missing = missing_per_column.sum()
#print("\nTotal missing values in DataFrame:", total_missing)

percent_missing = data.isnull().mean() * 100
print("\nPercent missing per column:\n", percent_missing.round(2))
```

```python
# Removing column "4b1. And are you satisfied with what you have?" due to high missing values
data_cleaned = data.drop(columns=["4b1. And are you satisfied with what you have?"])
```

```python
# Number of columns in the cleaned dataset
print("\nNumber of columns in cleaned dataset:", data_cleaned.shape[1])
```

```python
rows_missing_counts = data_cleaned.isnull().sum(axis=1)
# Show rows with many missing cells (e.g., 3 or more)
many_missing_rows = data_cleaned[rows_missing_counts >= 4]
print("\nNumber of rows with >=4 missing cells:", len(many_missing_rows))
many_missing_rows.head(10)  # preview first 10 such rows

```

```python
# Removing rows with more than 4 missing values
data_dropped_rows = data_cleaned.dropna(thresh=data_cleaned.shape[1] - 4)
print("\nShape after dropping rows with more than 4 missing values:", data_dropped_rows.shape)
```
